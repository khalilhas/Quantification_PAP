```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 22})
# Bethe law
density = 19.3 #g/cm3
Z = 79
A = 196.967
Em = np.linspace(1,100,num=100)# mean electron energy (keV) along the path
J = (9.76*Z + 58.5*pow(Z,-0.19))*pow(10,-3) # mean ionization potential (keV)
def dE_over_dx(Em):
    dEdx = -7.85*pow(10,4)*(Z*density/A*Em)*math.log(1.166*Em/J)# keV/cm
    return dEdx
dE_over_dx = np.vectorize(dE_over_dx)
dEdx = dE_over_dx(Em)
plt.figure(figsize=(12,12))
plt.plot(Em, dEdx)
plt.ylabel('dE/dx')
plt.xlabel('E(keV)')
print("Plot of the rate of energy loss with distance traveled (eV jnm) for gold a function of electron energy from the Bethe expression Equation (3.5), Goldstein book 1981.")
plt.show()
```
